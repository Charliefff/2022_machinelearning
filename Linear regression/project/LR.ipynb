```python
#  X (3, 30) Y(30, 1) W(30, 1) lr = 0.001
class LinearRegression:
    def __init__(self, learning_rate=0.001, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        
    # def gradient(self, X, Y):
    #     self.loss(X, Y)

    def loss(self, X, Y):
        return (((self.predict(X))-Y)**2)


    def train(self, X, Y):
        # 4*1
        self.weights = np.zeros((X.shape[1], 1))
        
        for i in range(self.n_iters):
            Predict = self.predict(X)
            dw = (2/(X.shape[0])*np.matmul(X.T, self.predict(X)-Y))
            self.weights -= self.lr*dw
            if i % 20 ==0:
                print("lr {} = {}".format(i, self.loss(X, Y)))
        return self.weights

    def predict(self, X):
        return (np.matmul(X, self.weights))
```

```python
x1, x2, x3, y = np.loadtxt("pizza_3_vars.txt", skiprows=1, unpack=True)
X = np.column_stack((np.ones(x1.size),x1, x2, x3))
Y = y.reshape(-1, 1)
regressor = LinearRegression(learning_rate=0.001, n_iters=300)
weights=regressor.train(X, Y)

# print("\nWeights: %s" % weights)
# print("\nA few predictions:")
# for i in range(5):
#     print("X[%d] -> %.4f (label: %d)" % (i, regressor.predict(X[i]), Y[i]))
```
