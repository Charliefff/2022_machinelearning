# 資料讀取

```python
import scipy.io

import numpy as np
import matplotlib.pyplot as plt
```

```python
org_data = scipy.io.loadmat('EEG_data\data_01.mat')

eeg_signal = org_data['data']
eeg_label = org_data['label']
```

# 畫圖

```python
print(eeg_label)
plt.figure()
plt.subplot(211)
plt.plot(eeg_signal[:, 0][2000:])
plt.subplot(212)
plt.plot(eeg_signal[:, 1])
```

# 濾波Example

```python
from scipy import signal

# 設定濾波器
sample_rate = 250
down_freq = 5. * 2 / sample_rate #頻率下限 2.5Hz
up_freq = 10. * 2 / sample_rate #頻率上限 50Hz
filt_range = [down_freq, up_freq]
b_p, a_p = signal.butter(6 , filt_range , 'bandpass')


# 濾波
fp1_signal = eeg_signal[:, 0]
fp2_signal = eeg_signal[:, 1]

fp1_signal = signal.filtfilt(b_p, a_p, fp1_signal)
fp2_signal = signal.filtfilt(b_p, a_p, fp2_signal)
```

```python
plt.figure()
plt.subplot(211)
plt.plot(fp1_signal[2000:])
plt.subplot(212)
plt.plot(fp2_signal)
```
