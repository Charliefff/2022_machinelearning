```python
# Imports
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss

def accuracy(y_true, y_pred):
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy

data = datasets.load_breast_cancer()
X, y = data.data, data.target

# y[y == 0] = -1

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)


sklearn_gbm = GradientBoostingClassifier(
    n_estimators=200, 
    learning_rate=0.1, 
    max_depth=1
)
sklearn_gbm.fit(X_train, y_train)
prob_train = sklearn_gbm.predict_proba(X_train)
sklearn_gbm_log_loss = log_loss(y_train, prob_train)
print(f"Scikit-learn GBM Log-Loss:{sklearn_gbm_log_loss:.15f}")
print(prob_train)
```

```python
print(y_test)
prob = sklearn_gbm.predict_proba(X_test)
y_pred = np.argmax(prob,axis=1)
print(y_pred)
acc = accuracy(y_test, y_pred)
print("Accuracy:", acc)
```
